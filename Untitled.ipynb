```python
import dicom
import numpy as np 
import array

import matplotlib.pyplot as plt
%matplotlib inline

from __future__ import print_function
from ipywidgets import interact
def ct_viewer(im):
    def plot_image(myslice):
        plt.imshow(im[myslice])#, cmap=plt.cm.gray)
    interact(plot_image, myslice=(0, im.shape[0]))
    return;
rd=dicom.read_file('rt_dose/rdb6e.dcm')
```

```python
print(rd)
print(rd.DoseGridScaling)
```

```python
pix=rd.pixel_array
ct_viewer(pix)
```

```python
from plotly.offline import init_notebook_mode, plot, iplot
from plotly.graph_objs import *
init_notebook_mode(connected=True)


plt.imshow(pix[:,12,:].T, origin='lower', cmap=plt.cm.Greys)
print(pix.shape)
profile=pix[69,8,:]
#profile=profile*4.0e-6
#profile=profile/1.8
profiledata=np.zeros((121, 2))

profiledata[:, 0]=np.arange(0, 121)*2.5
profiledata[:, 1]=profile

profiledata[:, 0]=profiledata[:, 0]-(profiledata[:, 0].max()/2)
profiledata[:, 1]=np.flipud(profiledata[:, 1])


data = Scatter(x = profiledata[:, 0],
               y = profiledata[:, 1],
               mode = 'lines+markers')

layout = dict(title='My scatter plot',
              font=dict(family='Droid Sans Mono',size=18), 
                )

fig = dict(data=[data], layout=layout)
iplot(fig)
```
