```python
import sys , os 
import numpy as np
import math as m
import pandas as pd

import plotly.offline as py


import plotly.graph_objs as go
py.offline.init_notebook_mode(connected=True)
g1="/home/crct/CRCT/JupyterNotebook/gammaIndex/9E/new/pdd9e30Gm_uncertainty.csv"
g2="/home/crct/CRCT/JupyterNotebook/gammaIndex/9E/new/pdd9e30Gw_uncertainty.csv"
g1u="/home/crct/CRCT/JupyterNotebook/gammaIndex/9E/new/pdd9e30Gm_U.csv"
g2u="/home/crct/CRCT/JupyterNotebook/gammaIndex/9E/new/pdd9e30Gw_U.csv"

g1Data=pd.read_csv(g1, sep=',')
g2Data=pd.read_csv(g2, sep=',')
g1Datau=pd.read_csv(g1, sep=',')
g2Datau=pd.read_csv(g2, sep=',')
#Create numpy ndarray from pandas dataFrames
g1x=g1Data['pas']
g1y=g1Data['value']
g2x=g2Data['pas']
g2y=g2Data['value']
g1xu=g1Datau['pas']
g1yu=g1Datau['value']
g2xu=g2Datau['pas']
g2yu=g2Datau['value']

# Plot profil or PDD with plotly      
traceg1 = go.Scatter(
    x = g1x,
    y = g1y,
    mode = 'lines', 
    name = 'Gate media', 
    line = dict(
        color = ('rgba(214, 39, 40, 100)')
    )
)

# Plot profil or PDD with plotly      
traceg2 = go.Scatter(
    x = g2x,
    y = g2y,
    mode = 'lines', 
    name = 'Gate water', 
    line = dict(
        color = '#7CBAE6'
    )
)

layout = dict(title='My scatter plot',
              font=dict(family='Droid Sans Mono',size=18), 
                )

datapdd = [traceg1, traceg2]

fig = go.Figure(data=datapdd, layout=layout)
py.iplot(fig)

```
