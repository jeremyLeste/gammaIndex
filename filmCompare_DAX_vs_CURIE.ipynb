```python
file_curie='/home/crct/CRCT/JupyterNotebook/data/gammaIndex/article/curie/'+energy+'e/'+heter+'/'+energy+'film'+heter+depth+'.csv'
file_dax='/home/crct/CRCT/JupyterNotebook/data/gammaIndex/dax/'+energy+'E/new/f'+heter+energy+'e'+depth+'.csv'
data1=pd.read_csv(file_curie, sep=',')
data2=pd.read_csv(file_dax, sep=',')
#print(data2)
```
