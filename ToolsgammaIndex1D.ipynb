```python
import sys , os 
import numpy as np
import math as m
import pandas as pd

import plotly.offline as py
import plotly.graph_objs as go
py.offline.init_notebook_mode(connected=True)
```

```python
#CourbeRef="testref.txt"
#Courbe="test.txt"
CourbeRef="mesure_9E_PDD.csv"
Courbe="gate_9E_PDD.csv"

df1=pd.read_csv(CourbeRef, sep=';')
df1


df2=pd.read_csv(Courbe, sep=';')
df2
```

```python
refX=df1['pas']
refY=df1['value']
gateX=df2['pas']
gateY=df2['value']

print type(refX[0])
print type(gateX[0])
print type(refY[0])
print type(gateY[0])

DTA=2 #mm
deltaDmax=0.02


deltaD=np.zeros(len(gateY))


gammaIndex=np.zeros(len(refX))
gammaIndexTEMP=np.zeros(len(gateX))
drTEMP=np.zeros(len(gateX))
dr=np.zeros(len(gateX))

for i in range(len(refX)):
    for j in range(len(gateX)):
        dr[j]=(refX[i]-gateX[j])
        deltaD[j]=refY[i]-gateY[j]
        gammaIndexTEMP[j]=m.sqrt(((deltaD[j]**2)/deltaDmax**2) + ((dr[j]**2)/DTA**2))
    gammaIndex[i]=np.min(gammaIndexTEMP)

print
print "GAMMA INDEX :"
print gammaIndex

nok=0.0

#a=0
#for i in range(len(gammaIndex)):
    
#    if gateY[i]>0.10:
#        a=a+1
#        if gammaIndex[i]<1:
#            nok=nok+1

for i in range(len(gammaIndex)):
        if gammaIndex[i]<1:
            nok=nok+1
            
                       

#print len(gammaIndex)
#passRate=(nok/a)*100
passRate=(nok/len(gammaIndex))*100

print
print('Pass Rate : ' +str(passRate) +' %')


trace1 = go.Scatter(
    x = refX,
    y = refY,
    mode = 'lines+markers', 
    name = 'mesure')

trace2 = go.Scatter(
    x = gateX,
    y = gateY,
    mode = 'lines+markers', 
    name = 'GATE')


xgamma=range(len(gammaIndex))
trace3 = go.Scatter(
    x = refX,
    y = gammaIndex,
    mode = 'lines+markers', 
    name = 'gamma index')
            
data = [trace1, trace2]
py.iplot(data, filename='scatter-mode')   


data1 = [trace3]
py.iplot(data1, filename='scatter-mode')   
```
